```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
class svm():
  def __init__(self,niter,lamda,lr):
     self.lr=lr
     self.niter=niter
     self.lamda=lamda
  def fit(self,x,y):
    self.m,self.n=x.shape
    self.x=x
    self.y=y
    self.w=np.zeros(self.n)
    self.b=0
    for i in range(self.niter):
      for yi,xi in enumerate(x):
        #cost function J= lamda*||w||^2+ summation (1,m) max(0,1-yi*(w*xi-b))
        #yi*(w*xi-b) >=1  ji= lamda*||w||^2
        #else ji= lamda*||w||^2+1-yi*(w*xi-b)
        if ( y[yi]*(np.dot(xi,self.w)-self.b) >=1):
          dw= 2*self.lamda*self.w
          db=0
        else:
          dw=2*self.lamda*self.w-np.dot(y[yi],xi)
          db= y[yi]
        self.w-= self.lr*dw
        self.b-=self.lr*db
  def predict(self,x):
    output=np.dot(x,self.w)-self.b
    return np.sign(output)
path ="/content/drive/MyDrive/data sets/Data.csv"
df=pd.read_csv(path)
X=df[['feature1','feature2']].values
Y=df['target'].values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=1/3,random_state=2)
svm1=svm(niter=1000,lamda=0.001,lr=0.01)
svm1.fit(X_train,Y_train)
y_predict=svm1.predict(X_test)
#print(y_predict,Y_test)
print("w1",svm1.w[0],"w2",svm1.w[1],"b",svm1.b)
c=np.sum(y_predict==Y_test)
print("accuracy", (c/len(Y_test))*100)
#2.572642243	3.97610927	1
```
