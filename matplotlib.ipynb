```python
import pandas as pd
import matplotlib.pyplot as plt
path="/content/drive/MyDrive/data/company_sales_data.csv"
x=pd.read_csv(path)
y=x['total_profit']
print(y)
plt.plot(y,ls='--',c='r',marker='o',mec='r',lw=3,label='Total profit')
plt.xlabel("month number")
plt.ylabel("profit in dollar")
plt.legend(loc="lower right")
plt.show()
```

```python
print(x)
y=x["month_number"]
print(y)
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
for vars in products:
  #print(x[vars])
  plt.plot(y,x[vars],label=vars)
plt.legend()
plt.show()
```

```python
print(x)
year_sales=[]
for sales in products:
  i=x[sales].sum()
  year_sales.append(i)
print(year_sales)
total_sales=sum(year_sales)
print(total_sales)
percentage=[]
for z in year_sales:
  percentage.append(z/total_sales)
print(percentage)
my_labels=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
plt.pie(percentage,labels=my_labels,autopct='%1.1f%%')
plt.show()
```
