Define a function to extract features from the given filename. You will have to use librosa library for this. You can vary the features however you like. The preferred features are mfcc, chroma and melspectogram.

here we take mean of feature  because the features extracted from librosa are of varying lengths due to differences in the audio file durations and the nature of the features themselves

```python
def extract_features(filename, list_of_features):
  features=[]
  audio,sample_rate=librosa.load(filename)
  for feature in list_of_features:
        if feature == 'mfcc':
            mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13)
            mfccs_mean = np.mean(mfccs.T, axis=0)
            #print("mfcc",np.shape(mfccs_mean))
            features.append(mfccs_mean)


        elif feature == 'chroma':
            chroma = librosa.feature.chroma_stft(y=audio, sr=sample_rate)
            chroma_mean = np.mean(chroma.T, axis=0)
            #print("chroma",np.shape(chroma_mean))
            features.append(chroma_mean)

# feature == 'melspectrogram':
        else:
            mel_spec = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_fft=2048, hop_length=512)
            mel_spec_mean = np.mean(mel_spec.T, axis=0)
            #print("mel",np.shape(mel_spec_mean ))
            features.append(mel_spec_mean)


  return np.hstack(features)
```
