**inverse of a matrix**

```python
import numpy as np
import math
def inverse(arr,n):
  cofactor=np.zeros((arr.shape))
  for i in range(n):
    for j in range(n):
      rows=[i]
      columns=[j]
      remaining=np.delete(arr,rows,axis=0)
      remaining=np.delete(remaining,columns,axis=1)
      sign=(-1) ** (i + j)
      cofactor[i,j]=sign * (np.linalg.det(remaining))
  adj=cofactor.T
  return adj
arr=np.array(np.random.randint(1,20,size=(3,3)))
print("arr\n",arr)
n=arr.shape[0]
if (arr.shape[0]==arr.shape[1]):
  print("it is square matrix")
  deter=np.linalg.det(arr)
  if(deter==0):
    print("inverse doesn't exist")
  else:
    print("inverse exist")
  arr_inverse=inverse(arr,n) /deter
  print("calculated arr_ inverse matrix is \n",arr_inverse)
  identity=np.matmul(arr,arr_inverse)
  print("arr*arr_inverse is identity \n",identity)
  expected_arr_inverse=np.linalg.inv(arr)
  print("inverse matrix using linalg.inv \n",expected_arr_inverse)
  equal= np.allclose(expected_arr_inverse,arr_inverse)
  print(equal)

else:
  print("it is not a square matrix \n we cannot find inverse  to non square matrix")
```
